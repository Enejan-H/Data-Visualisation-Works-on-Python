## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir

```python
m = np.linspace(0, 10, 11)  # 0'dan 10'a kadar eşit aralıklı 11 sayı
c = 3 * 10**8               # Işık hızı (m/s)
E = m * c**2                # Enerji hesaplaması

# Grafik oluşturma
plt.plot(m, E, color="green", lw=3)  # Enerji-madde grafiği
plt.title("E = mc^2")                # Grafik başlığı
plt.xlabel("M in g")                 # x-ekseni etiketi
plt.ylabel("E in J")                 # y-ekseni etiketi
plt.xlim(0, 10)                      # x-ekseni sınırları
plt.yscale("log")                    # y-eksenini logaritmik ölçeğe ayarlama
plt.grid(which="both", axis="y")     # y-ekseninde ızgara çizgilerini ayarlama
plt.show()
```

## Aşağıdaki değişkenlere göre ilgili platlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
plt.figure(figsize=(5.5,4))

plt.plot(labels, july16_2007, "SteelBlue", label="july16_2007")
plt.plot(labels, july16_2020, "orange", label="july16_2020")

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(5.5,4))

plt.plot(labels, july16_2007, "SteelBlue", label="july16_2007")
plt.plot(labels, july16_2020, "orange", label="july16_2020")

plt.legend(bbox_to_anchor = (1, 0.5, 0.35, 0))

plt.show()
```

## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python

fig, ax = plt.subplots(2, 1)
ax[0].plot(labels, july16_2007, "SteelBlue", lw=2)
ax[0].set_title("july16_2007", fontsize=9)
ax[1].plot(labels, july16_2020, "SteelBlue", lw=2)
ax[1].set_title("july16_2020", fontsize=9)


plt.tight_layout()
plt.show()
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python

labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]


fig, ax1 = plt.subplots(figsize=(10, 6))


color = 'blue'
ax1.set_xlabel('')
ax1.set_ylabel('July 16, 2007', color=color)
ax1.plot(labels, july16_2007, color=color, linestyle="-")
ax1.tick_params(axis='y', labelcolor=color)

# Creating a twin axis on the right side
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('July 16 2020', color=color)
ax2.plot(labels, july16_2020, color=color,linestyle="-")
ax2.tick_params(axis='y', labelcolor=color)

# Adding labels and title

# Display the plot
plt.show()
```
