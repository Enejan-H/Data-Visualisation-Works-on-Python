## 2: Plot 12 most dangerous cities

```python
df.city.value_counts().head(12)
```

```python
df.city.value_counts().head(12).values
```

```python
df.city.value_counts().head(12).index
```

```python
plt.figure(figsize=(7,4))

sns.barplot(x=df.city.value_counts().head(12).index,
            y=df.city.value_counts().head(12).values)

plt.title("Most Dangerous Cities",c="red", fontsize=10)
plt.xticks(rotation=45, x ='right', fontsize=8, color='b')
plt.tight_layout()
```

```python
# with seaborn
```

```python
df.city.value_counts().head(12)
```

```python
df.city.value_counts().head(12).index
```

```python

plt.figure(figsize=(6,3))

sns.countplot(x="city", 
              data=df[df["city"].isin(df.city.value_counts().head(12).index)], 
              order=df.city.value_counts().head(12).index, 
              width=0.8)

plt.title("Most Dangerous Cities", color='b')
plt.xticks(rotation=45)
plt.show()
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states

```python
df.state.value_counts().head(20)
```

```python
plt.figure(figsize=(5,3.5))

sns.barplot(x=df.state.value_counts().head(20).index, 
            y=df.state.value_counts().head(20).values)

plt.title("Most Dangerous States",c="b", fontsize=10)
plt.xticks(rotation=45, x ='right', fontsize=8)
plt.tight_layout()
```

```python
# with seaborn barplot
```

```python
plt.figure(figsize=(5.5,3.5))

sns.countplot(x="state", 
              data=df[df["state"].isin(df.state.value_counts().head(20).index)], 
              order=df.state.value_counts().head(20).index, 
              width=0.8)

plt.title("Most Dangerous States", color='b')
plt.xticks(rotation=45);
```
